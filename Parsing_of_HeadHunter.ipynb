```python
seniority = ["Junior", "Lead", "Middle", "Senior"]
seniority_list = []
a = -1
for i in positions:
    ok = 0
    a += 1
    loc = i.split()
    for j in loc:
        if j in seniority or j == "Менеджер":
            ok = 1
            seniority_list.append(j)
            break
        else:
            loc = ""
            for f in range(0, len(j)):
                if j[f] in ['J', 'L', 'M', 'S'] and len(loc) == 0:
                    loc += j[f]
                elif len(loc) > 0:
                    loc += j[f]
                    if (len(loc) == 4 or len(loc) == 6) and loc in seniority:
                        seniority_list.append(loc)
                        ok = 1
                        break
    if ok == 0:
        if wages_list[a] < 12000:
            seniority_list.append("Junior")
        elif 12000 <= wages_list[a] < 21000:
            seniority_list.append("Middle")
        else:
            seniority_list.append("Senior")
            
```
